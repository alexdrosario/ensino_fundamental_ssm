```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pysal as ps
import geopandas as gp
from nupis import *
import palettable
%matplotlib inline

atlas = pd.read_csv('atlas.csv')
```

## a fazer

- preencher cidades em braco no brasil
- usar as mesmas faixas para todos os mapas

```python
br = gp.read_file('shapefiles/BRMUE250GC_SIR.shp')
br.geometry = br.geometry.simplify(0.001)
br.rename(columns={'CD_GEOCMU': 'mun'}, inplace=True)
br['mun'] = br.mun.astype(int)
```

```python
br = pd.merge(br, atlas, on='mun')
ax = br.plot(column='e_anosestudo', scheme='equal_interval', k=4, cmap='viridis',
             linewidth=0, figsize=(15,15), legend=True)
ax.set_axis_off();
```
