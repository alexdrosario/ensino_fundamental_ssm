```python
# abre o arquivo escolas (censo escolar - 2015)
link = '../../../../Downloads/inep/microdados_censo_escolar_2015/DADOS/'
esc = pd.read_csv(link+'ESCOLAS.CSV', encoding='latin_1', sep= '|', low_memory=False)

# subseleciona ssm
esc = esc.query('CO_MESORREGIAO == 3110')

# passa o nome das colunas para caixa baixa
simplificar_colunas(esc)
# retira acentos e espaços dos nomes das escolas
simplificar_nomes(esc, 'no_entidade')

# dimensões do dataframe
esc.shape

# para conseguir algumas informações sobre os municípios.
# vamos fundir o dataframe das escolas com o dataframe atlas, 
# que possui dezenas de variáveis sobre os municípios

# renomeia variável para coincidir com o atlas
esc.rename(columns={'co_municipio': 'mun'}, inplace=True)

# carrega o atlas e seleciona algumas variáveis
atlas = pd.read_csv('dados/atlas.csv')
atlas = atlas.loc[:,['nome_mun', 'mun', 'micro', 'nome_micro', 'popul', 'porte']]

# funde o dataframe das escolas com o atlas
esc = pd.merge(esc, atlas, on='mun')
```

```python
esc.shape
```

```python
# mostra variáveis disponíveis 
# as informações sobre elas estão em:
# microdados_censo_escolar_2015/DADOS/ANEXOS/ANEXO I.xls
esc.columns
```

```python
# escolas por dependência administrativa
# 1 - federal 
# 2 - estadual
# 3 - municipal
# 4 - privada
esc.tp_dependencia.value_counts()
```

```python
esc.tp_dependencia.value_counts().plot.bar();
```

```python
# escolas por dependência administrativa
# pequenoI: até 20 mil 
# pequenoII: entre 20 e 50 mil
# medioI: entre 50 e 100 mil
# medioII: entre 100 e 500 mil 

esc.groupby('porte').tp_dependencia.value_counts()
```

```python
esc.groupby('porte').tp_dependencia.value_counts().plot.barh();
```

```python
x = esc.groupby('porte').tp_dependencia.value_counts()
xx= pd.DataFrame(x)
xx.columns = ['valor']
xx = xx.reset_index()
xx = xx.pivot_table(index='porte', columns='tp_dependencia', values='valor')
```

```python
xx.plot.barh(stacked=True);
```

```python
xx.transform(lambda x: x / x.sum() * 100, axis=1).plot.barh(stacked=True)

plt.legend(title='tipo de escola', labels=['federal', 'estadual', 'municipal', 'privada'], 
           loc='upper center', bbox_to_anchor=(1.1, 1))

plt.xlim(0,100);
```

```python
import pandas as pd
```

```python
esc = pd.read_csv('microdados_saeb_2015/DADOS/TS_ESCOLA.csv', encoding='latin1')
esc.columns = esc.columns.str.lower()
esc.rename(columns={'id_municipio': 'mun'}, inplace=True)

#criando o banco de dados da mesorregião do SM
atlasmesmg = pd.read_csv('atlas.csv').query('meso == 3110')
mesmg = pd.merge(atlasmesmg.iloc[:,:7], esc, on='mun')

#criando o banco de dados de MG
atlasmg = pd.read_csv('atlas.csv').query('uf == 31')
ssmg = pd.merge(atlasmg.iloc[:,:7], esc, on='mun')

#criando o banco de dados do Brasil
atlasbr = pd.read_csv('atlas.csv')
ssbr = pd.merge(atlasbr.iloc[:,:7], esc, on='mun')

# substituindo NAs pela média
mesmg['media_5ef_lp'] = mesmg.media_5ef_lp.fillna(mesmg.media_5ef_lp.mean())
mesmg['media_5ef_mt'] = mesmg.media_5ef_mt.fillna(mesmg.media_5ef_mt.mean())
ssmg['media_5ef_lp'] = ssmg.media_5ef_lp.fillna(ssmg.media_5ef_lp.mean())
ssmg['media_5ef_mt'] = ssmg.media_5ef_mt.fillna(ssmg.media_5ef_mt.mean())
ssbr['media_5ef_lp'] = ssbr.media_5ef_lp.fillna(ssbr.media_5ef_lp.mean())
ssbr['media_5ef_mt'] = ssbr.media_5ef_mt.fillna(ssbr.media_5ef_mt.mean())
```

```python
pd.options.display.max_seq_items = 500
```

```python
ssbr.shape
```

```python
#cálculo da média das notas de língua portuguesa
media_5ef_lp_mesmg = mesmg.media_5ef_lp.mean()
media_5ef_lp_mg = ssmg.media_5ef_lp.mean()
media_5ef_lp_br = ssbr.media_5ef_lp.mean()

#cálculo da média das notas de matemática
media_5ef_mt_mesmg = mesmg.media_5ef_mt.mean()
media_5ef_mt_mg = ssmg.media_5ef_mt.mean()
media_5ef_mt_br = ssbr.media_5ef_mt.mean()
```

```python
print ('Média Prova Brasil do 5º ano em língua portuguesa')
print ('Mesorregião SM:',round(media_5ef_lp_mesmg,2))
print ('Estado de MG..:',round(media_5ef_lp_mg,2))
print ('Brasil........:',round(media_5ef_lp_br,2))
print ('')
print ('Média Prova Brasil do 5º ano em matemática')
print ('Mesorregião SSM:',round(media_5ef_mt_mesmg,2))
print ('Estado de MG...:',round(media_5ef_mt_mg,2))
print ('Brasil.........:',round(media_5ef_mt_br,2))
```
