## pib dividido por setores

```python
mesopib = pibmun.groupby('meso')['vab_agro', 'vab_ind', 'vab_serv', 'vab_adm', 'vab_total', 'impostos', 'pib', 'pop'].sum()
```

```python
mesopib['agro'] = round(mesopib['vab_agro'] / mesopib['pib'], 2)
mesopib['ind'] = round(mesopib['vab_ind'] / mesopib['pib'], 2)
mesopib['serv'] = round(mesopib['vab_serv'] / mesopib['pib'], 2)
mesopib['adm'] = round(mesopib['vab_adm'] / mesopib['pib'], 2)
mesopib['vab'] = round(mesopib['vab_total'] / mesopib['pib'], 2)
mesopib['imp'] = round(mesopib['impostos'] / mesopib['pib'], 2)
mesopib['pibpc'] = (mesopib['pib'] * 1000) / mesopib['pop'] 
```

Minas Gerais continua em melhor situação em 2015 nessa faixa etária. Já nas outras idades, São Paulo é que se encontra em melhor situação, especialmente no ensino médio (faixa etária de 15 a 17 anos). São Paulo apresenta uma ampla vantagem, com cerca de 67,7%, em 2004, aumentando para 77,2%, em 2015. Já em Minas Gerais, esse percentual é de 49,8%, em 2004, aumentando para 61%, em 2015, um número menor do que São Paulo em 2004, ou seja, mesmo melhorando de 2004 para 2015, Minas Gerais ainda não alcançou o percentual que São Paulo possuía em 2004. Na faixa etária de 11 a 14 anos, que são os anos finais do ensino fundamental, Minas Gerais se igualou a São Paulo em 2015.

   Como foi visto, o estado de Minas Gerais possui, em todos os níveis de ensino, maior número de alunos na rede pública, com exceção apenas para o ensino superior. Porém, de 2004 para 2015, o estado de Minas Gerais teve uma melhora, alcançando ou ao menos ficando bem próximo dos índices de São Paulo.
```python
quartil91 = []
quartil00 = []
quartil10 = []

for linha in ed.flfund91:
    if linha <= 83.45:
        quartil91.append('1')
    elif 83.45 < linha <= 88.18:
        quartil91.append('2')
    elif 88.18  < linha <= 92.92:
        quartil91.append('3') 
    else:
        quartil91.append('4')

ed['quartil91'] = quartil91


for linha in ed.flfund00:
    if linha <= 83.45:
        quartil00.append('1')
    elif 83.45 < linha <= 88.18:
        quartil00.append('2')
    elif 88.18  < linha <= 92.92:
        quartil00.append('3') 
    else:
        quartil00.append('4')

ed['quartil00'] = quartil00


for linha in ed.flfund10:
    if linha <= 83.45:
        quartil10.append('1')
    elif 83.45 < linha <= 88.18:
        quartil10.append('2')
    elif 88.18  < linha <= 92.92:
        quartil10.append('3') 
    else:
        quartil10.append('4')

ed['quartil10'] = quartil10

ed['quartil91'] = ed.quartil91.astype(int)
ed['quartil00'] = ed.quartil00.astype(int)
ed['quartil10'] = ed.quartil10.astype(int)
```
